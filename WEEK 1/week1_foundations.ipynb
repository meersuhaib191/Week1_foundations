```python
n1 = int(input("Enter any number = "))
sum = 0
temp = n1
digits = len(str(n1))

while temp > 0 :
  dig = temp%10
  sum= sum + dig**digits
  temp = temp // 10

if sum == n1 :
   print("armstrong number")
else :
  print("not an armstrong number")
```
