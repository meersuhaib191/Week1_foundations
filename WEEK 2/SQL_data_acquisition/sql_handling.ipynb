```python
import sqlite3
import pandas as pd


conn = sqlite3.connect("car_sales.db")
cursor = conn.cursor()

df = pd.read_csv("used_cars.csv")

df.to_sql("used_cars", conn, if_exists="replace", index=False)

df_sql = pd.read_sql("SELECT * FROM used_cars LIMIT 10;", conn)
df_sql
```

```python
ford_cars = pd.read_sql("SELECT * FROM used_cars WHERE brand = 'Ford';", conn)
ford_cars
```

```python
model = int(input("Enter the model year to filter cars: "))

query = f"SELECT * FROM used_cars WHERE model_year = {model};"
result = pd.read_sql(query, conn)
result
```

```python
# Count NULLs in each column
nulls = pd.read_sql("""
SELECT
    SUM(CASE WHEN brand IS NULL OR brand = '' THEN 1 ELSE 0 END) AS brand_null,
    SUM(CASE WHEN model IS NULL OR model = '' THEN 1 ELSE 0 END) AS model_null,
    SUM(CASE WHEN model_year IS NULL THEN 1 ELSE 0 END) AS model_year_null,
    SUM(CASE WHEN milage IS NULL THEN 1 ELSE 0 END) AS milage_null,
    SUM(CASE WHEN fuel_type IS NULL OR fuel_type = '' THEN 1 ELSE 0 END) AS fuel_type_null,
    SUM(CASE WHEN price IS NULL THEN 1 ELSE 0 END) AS price_null,
    SUM(CASE WHEN accident IS NULL OR accident = '' THEN 1 ELSE 0 END) AS accident_null
FROM used_cars;
""", conn)

nulls
```

```python
missing_fuel = pd.read_sql("""
SELECT *
FROM used_cars
WHERE fuel_type IS NULL OR fuel_type = '';
""", conn)

missing_fuel
```

```python
missing_accident_report = pd.read_sql("""
SELECT *
FROM used_cars
WHERE accident IS NULL OR accident = '';
""", conn)

missing_accident_report
```

```python
# Step 1: Create a cursor
cursor = conn.cursor()

cursor.execute("""
UPDATE used_cars
SET accident = 'Inspection Required'
WHERE accident IS NULL OR accident = '';
""")

conn.commit()
cursor.close()
```
